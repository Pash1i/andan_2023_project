```python
import numpy as np
import pandas as pd

import scipy.stats as sts
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Важный_файл_3.csv', sep=',')
df
```

# Графики: столбчатые диаграммы

```python
plt.figure(figsize=(20,10))
plt.bar(df['date'], df['n_views'], width=0.8, color = "crimson")
plt.xlabel("Дата")
plt.ylabel("Количество просмотров, млн")
plt.title("Распределение количества просмотров твита в зависимости от даты")
ticks = list(df['date'])
plt.xticks([ticks[i] for i in range(len(ticks)) if i % 100 == 0], rotation='vertical')
plt.show()
```
